```python
def UD(diff):
    ups = 0
    downs = 0
    if diff > 0:
        ups = diff
    elif diff < 0:
        downs = abs(diff)

    return pd.Series([ups, downs])
```

```python
df.btc["diff"].apply(UD)
```

```python
df.btc[["U", "D"]] = df.btc["diff"].apply(UD)
df.btc
```

```python
df.btc["AU"] = df.btc["U"].rolling(14).mean()
df.btc["AD"] = df.btc["D"].rolling(14).mean()
df.btc
```

```python
df.btc["RS"] = df.btc["AU"] / df.btc["AD"]
df.btc["RSI"] = df.btc["RS"] / (1 + df.btc["RS"])
df.btc
```

```python
df.btc['RSI_signal'] = df.btc["RSI"].rolling(6).mean()
df.btc
```

```python
# 20일 종가 표준편차
# UpperB = 종가의 이동평균 + (표준편차 * 2)
# LowerB = 종가의 이동평균 - (표준편차 * 2)
df.btc["stddev"] = df.btc["Adj Close"].rolling(20).std()
df.btc["UpperB"] = df.btc["ma20"] + (df.btc["stddev"] * 2)
df.btc["LowerB"] = df.btc["ma20"] - (df.btc["stddev"] * 2)
df.btc
```

```python
# 볼린저밴드(BollingerBand): "UpperB", "Adj  Close", , "ma20"LowerB" 시각화
df.btc[["UpperB", "Adj Close", "ma20", "LowerB"]].plot(figsize=(20,10))
plt.fill_between(df.btc.index, df.btc.LowerB, df.btc.UpperB, alpha=0.2)
```

```python
# %b = 종가 - 하위밴드 / 상위밴드 - 하위밴드
df.btc["PercentB"] = (df.btc["Adj Close"] - df.btc["LowerB"]) / (df.btc["UpperB"] - df.btc["LowerB"])
```

```python
# %b 시각화
plt.figure(figsize=(20, 10))
plt.bar(df.btc.index, df.btc["PercentB"])
plt.axhline(1, c="r", lw=1)
plt.axhline(0.5, c="b", lw=1)
plt.axhline(0, c="k", lw=1)
```

```python
# 12일, 26일 지수이동평균
short = 12
long = 26
df.btc["ema_short"] = df.btc["Adj Close"].ewm(short).mean()   # 12일간의 지수 이동평균
df.btc["ema_long"] = df.btc["Adj Close"].ewm(long).mean()     # 26일간의 지수 이동평균
```

```python
# "Adj Close", "ema_short", "ema_long" 을 시각화 합니다.
df.btc[["Adj Close", "ema_short", "ema_long"]].plot(figsize=(15,5))
```

```python
# MACD = 단기 지수이동평균 - 장기 지수이동평균
df.btc["MACD"] = df.btc["ema_short"] - df.btc["ema_long"]
```

```python
df.btc
```

```python
# MACD 값 시각화
df.btc["MACD"].plot(figsize=(15, 5))
plt.axhline(0)
```

```python
# MACD 9일 이동평균값
df.btc["signal"] = df.btc["MACD"].ewm(9).mean()
df.btc[["MACD", "signal"]].plot(figsize=(15, 5))
plt.axhline(0)
```

```python
# MACD oscillator 값
df.btc["macdhist"] = df.btc["MACD"] - df.btc["signal"]
plt.figure(figsize=(15, 5))
plt.bar(df.btc.index, df.btc.macdhist)
```
